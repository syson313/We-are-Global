```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, recall_score, precision_score, f1_score

X_train = pd.read_csv('../data/heart_X_train.csv')
X_test = pd.read_csv('../data/heart_X_test.csv')
y_train = pd.read_csv('../data/heart_y_train.csv')
y_test = pd.read_csv('../data/heart_y_test.csv')

lr = LogisticRegression()
X_train_column = X_train[['Smoking_1','Obesity_1','Alcohol Consumption_1','Previous Heart Problems_1','Medication Use_1','BloodPressure_1']]
X_train['av'] = X_train_column.sum(axis=1) / 6
X_train['av']

X_test_column = X_test[['Smoking_1','Obesity_1','Alcohol Consumption_1','Previous Heart Problems_1','Medication Use_1','BloodPressure_1']]
X_test['av'] = X_test_column.sum(axis=1) / 6
X_test['av']

lr.fit(X_train['av'].values.reshape(-1, 1), y_train['Heart Attack Risk'])

train_score = lr.score(X_train['av'].values.reshape(-1, 1), y_train['Heart Attack Risk'])
test_score = lr.score(X_test['av'].values.reshape(-1, 1), y_test['Heart Attack Risk'])

print("Training score:", train_score, "Testing score:", test_score)

input_av = np.array([0.63888889]).reshape(-1, 1)
predicted_class = lr.predict(input_av)
print("Predicted class:", predicted_class[0])
predicted_probabilities = lr.predict_proba(input_av)
print("Predicted probabilities:", predicted_probabilities)
print(f"Probability of class 0: {predicted_probabilities[0][0]}")
print(f"Probability of class 1: {predicted_probabilities[0][1]}")

from sklearn.linear_model import Ridge

ridge = Ridge(
    alpha=10,
    random_state=1)

ridge.fit(X_train, y_train)

y_train_pred_ridge = ridge.predict(X_train)
y_test_pred_ridge = ridge.predict(X_test)

from sklearn.metrics import mean_squared_error

print(f'Training R^2: {ridge.score(X_train, y_train)}')
print(f'Test R^2: {ridge.score(X_test, y_test)}')
print(f'Train RMSE: {mean_squared_error(y_train, y_train_pred_ridge, squared=False)}')
print(f'Test RMSE: {mean_squared_error(y_test, y_test_pred_ridge, squared=False)}')

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def logistic_regression(data, learning_rate=0.02, epochs=1000):
    w = 0.0
    b = 0.0
    for i in range(1, epochs + 1):
        for x, y in data:
            w_difference = x * (sigmoid(w * x + b) - y)
            b_difference = sigmoid(w * x + b) - y

            w -= learning_rate * w_difference
            b -= learning_rate * b_difference

        if i % 100 == 0:
            print(f"Epoch {i}: w = {round(w, 3)}, b = {round(b, 3)}")

    return w, b

# 데이터 준비
x_data = X_test['av'].values
y_data = y_test['Heart Attack Risk'].values

# (x, y) 형태로 데이터 합치기
data = np.column_stack((x_data, y_data))

# 로지스틱 회귀 수행
w, b = logistic_regression(data=data)

# 시각화
plt.figure(figsize=(10, 6))

# 실제 데이터 산점도
plt.scatter(x_data, y_data, label="Actual Data", color="blue", alpha=0.6, edgecolor="k")

# 로지스틱 회귀 함수
x_range = np.linspace(x_data.min(), x_data.max(), 500)
y_pred = sigmoid(w * x_range + b)
plt.plot(x_range, y_pred, color="red", label="Logistic Regression Curve")

# 그래프 꾸미기
plt.title("Logistic Regression: Heart Attack Risk vs AVERAGE")
plt.xlabel("AVERAGE")
plt.ylabel("Heart Attack Risk")
plt.legend()
plt.grid(True)
plt.show()
```
